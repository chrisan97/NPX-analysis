```python
# Set the base_folder and the folder that holds the spikeglx data
# spikeglx_folder should change for every recording that you perform
base_folder = Path('D:/NPX_data/DATA')

spikeglx_folder = base_folder / '05242024-NTQP4_g0/05242024-NTQP4_g0'
```

```python
# we do not load the sync channel, so the probe is automatically loaded
raw_rec = si.read_spikeglx(spikeglx_folder, stream_name='imec0.ap', load_sync_channel=False)
raw_rec
```

```python
# Here we are loading in the data and doing any necessary processing
rec1 = si.highpass_filter(raw_rec, freq_min=400.)
#bad_channel_ids, channel_labels = si.detect_bad_channels(rec1)
#rec2 = rec1.remove_channels(bad_channel_ids) # Here we remove the bad channels 
#print('bad_channel_ids', bad_channel_ids)

rec3 = si.phase_shift(rec1)
rec4 = si.common_reference(rec3, operator="median", reference="global") # We are using a common reference filter here; this is really not necessary as catGT/kilosort will take care of this
rec = raw_rec
rec
```

```python
# we can estimate the noise on the scaled traces (microV) or on the raw one (which is in our case int16).
noise_levels_microV = si.get_noise_levels(rec4, return_scaled=True)
noise_levels_int16 = si.get_noise_levels(rec4, return_scaled=False)
```

```python
# Set the parameters for kilosort 2.5
params_kilosort2_5  = si.get_default_sorter_params('kilosort2_5')
params_kilosort2_5['car'] = False
params_kilosort2_5['delete_recording_dat'] = True   # delete temporary files (recording.dat) 
params_kilosort2_5['save_rez_to_mat'] = True    # save the spike structure (rez.mat)
params_kilosort2_5['delete_tmp_files'] = False
params_kilosort2_5
```

```python
ks2_5_path = os.path.join('D:\\','Kilosort-2.5.2')

si.Kilosort2_5Sorter.set_kilosort2_5_path(ks2_5_path)
```

```python
print(f'starting kilosort2_5...')

ks2_5_sorter = si.run_sorter('kilosort2_5', rec, output_folder=spikeglx_folder/'kilosort2.5_output',remove_existing_folder=True, verbose=True, **params_kilosort2_5)
```

```python
analyzer = si.create_sorting_analyzer(ks2_5_sorter, rec, sparse=True, format="memory")
analyzer
```

```python
job_kwargs = dict(n_jobs=40, chunk_duration='1s', progress_bar=True)
analyzer.compute("random_spikes", method="uniform", max_spikes_per_unit=500)
analyzer.compute("waveforms",  ms_before=1.5,ms_after=2., **job_kwargs)
analyzer.compute("templates", operators=["average", "median", "std"])
analyzer.compute("noise_levels")
analyzer.compute("correlograms")
analyzer.compute("unit_locations")
analyzer.compute("spike_amplitudes", **job_kwargs)
analyzer.compute("template_similarity")
analyzer.compute('principal_components', n_components = 5, mode="by_channel_local")
analyzer
```

```python
analyzer_saved = analyzer.save_as(folder=spikeglx_folder/'kilosort2.5_output' / "analyzer", format="binary_folder")
analyzer_saved
```

```python
metric_names=['firing_rate', 'presence_ratio', 'snr', 'isi_violation', 'amplitude_cutoff']


# metrics = analyzer.compute("quality_metrics").get_data()
# equivalent to
metrics = si.compute_quality_metrics(analyzer, metric_names=metric_names)

metrics
```

```python
analyzer = si.load_sorting_analyzer(folder =spikeglx_folder/'kilosort2.5_output' / "analyzer")
```

```python
job_kwargs = dict(n_jobs=12, chunk_duration='1s', progress_bar=True)
si.export_to_phy(analyzer, output_folder = spikeglx_folder/'kilosort2.5_output' / "SI_phy", compute_pc_features=True, compute_amplitudes=True, copy_binary=False,**job_kwargs)
```
